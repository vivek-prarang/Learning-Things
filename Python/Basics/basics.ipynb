Basics OF Python

Python is a high-level, interpreted programming language known for its simplicity and readability. It was created by Guido van Rossum and first released in 1991. Python supports multiple programming paradigms, including procedural, object-oriented, and functional programming.

Get Started With Python:
1. Python is essy to learn and has a large community, making it a great choice for beginners, as well as experienced developers.
2. But the internly python bit diffent from other languages. Variable and memory management is handled automatically, and it has a dynamic type system, which means you don't need to declare variable types explicitly.
3. Garbage collection is also built-in, which helps to manage memory efficientlty. it use reference counting and cyclic garbage collection to free up memory that is no longer in use.
4. Python variables are not like containers that hold data, but rather they are references to objects in memory. 
5. When you assign a value to a variable, you are actually creating an object in memory and the variable is referencing that object.
